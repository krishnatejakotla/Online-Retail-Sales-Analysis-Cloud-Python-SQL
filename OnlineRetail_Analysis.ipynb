```python
# Check basic info about dataset
df.info()

# Check missing values in each column
df.isnull().sum()
```

```python
# Drop rows where CustomerID is missing
df = df.dropna(subset=['CustomerID'])

# Reset index
df = df.reset_index(drop=True)

print("Shape after dropping missing CustomerIDs:", df.shape)
```

```python
# Remove canceled orders (InvoiceNo starting with 'C')
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]

print("Shape after removing canceled transactions:", df.shape)
```

```python
# Remove negative Quantity values
df = df[df['Quantity'] > 0]

print("Shape after removing negative quantities:", df.shape)
```

```python
# Save the cleaned dataset
df.to_csv('OnlineRetail_Cleaned.csv', index=False)

print("Cleaned data saved successfully.")
```

```python
# Connect to a new SQLite database file
conn = sqlite3.connect('retail_data.db')

# Push the cleaned dataframe into a SQL Table
df.to_sql('OnlineRetail', conn, if_exists='replace', index=False)

print("Data inserted into SQLite database successfully!")
```
